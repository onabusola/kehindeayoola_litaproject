```python
df['study_hours_per_week'].value_counts()
```

```python
# Check for missing values in 'study_hours_per_week'
missing_values = df['study_hours_per_week'].isnull().sum()
print(f"Number of missing values in 'study_hours_per_week': {missing_values}")

# Check the data type of 'study_hours_per_week'
data_type = df['study_hours_per_week'].dtype
print(f"Data type of 'study_hours_per_week': {data_type}")

# Display descriptive statistics for the column
display(df['study_hours_per_week'].describe())
```

```python
# Rename the column 'study_hours_per_week' to 'study_hours'
df = df.rename(columns={'study_hours_per_week': 'study_hours'})

# Display the updated column names to confirm the change
print("Updated column names:")
print(df.columns)
```

```python
# Check for and count duplicate rows
duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")
```

```python
df.isnull ().sum()
```

```python
mean= df ['study_hours'].mean()
```

```python
print(f'The mean study hours per week is: {mean}')
```

```python
study_hours_final_grade_mean = df.groupby('study_hours')['final_grade'].mean()
display(study_hours_final_grade_mean)
```
